Naive Bayes algorithm is a supervised learning algorithm, which is based on Bayes theorem and used for solving classification problems.

It is mainly used in txt classification that includes a high-dimensional training dataset.

Naive bayes classifier is one of the simple and most effective classification algorithms which helps is building the fast machine learning models that can make predictions.

It is a probalistic classifier,which means it predicts on the basis of the probability of an object.

Some popular examples of naive Bayes algorithm are spam filteration,Sentimental Analysis, and classifying articles.

# Implementation of Naive Bayes

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Social_network_ads.csv")
data.head()
```

```python
data.info()
```

```python
data.describe().T
```

```python
data.isna().sum()
```

```python
X=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
```

```python
X.shape
```

```python
y.shape
```

# Splitting Dataset

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
print(X_train.shape)
X_train
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
X_train
```

```python
from sklearn.naive_bayes import GaussianNB
classifier=GaussianNB()
classifier.fit(X_train,y_train)
```

```python
classifier.predict(sc.transform([[30,87000]]))
```

Here '0' mean customer will not purchased the car

# Predicting on test dataset result

```python
y_pred=classifier.predict(X_test)
y_pred
```

# We can see the difference between real values of test dataset and Prediction values

```python
df=pd.DataFrame({
    "Y_test":y_test,
    "Y-pred":y_pred
})
```

```python
df
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
acc=accuracy_score(y_test,y_pred)
```

```python
print("Confusion matrix ",cm)
```

```python
print("The accuracy score of KNN model:",acc)
```
